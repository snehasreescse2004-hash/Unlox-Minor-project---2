```python
# TASK - 1
#Import Libraries
import pandas as pd
import numpy as np
```

```python
#Upload and Read the CSV File
df = pd.read_csv("Data set for DADS June.csv")
```

```python
#Display Basic Information
print(df.head())

print(df.info())

print(df.shape)
```

```python
#Remove Duplicate Rows
duplicate_rows = df.duplicated().sum()

print("Duplicate Rows :", duplicate_rows)

df = df.drop_duplicates()

print("Shape After Removing Duplicates :", df.shape)
```

```python
#Convert Date Column
df["Date"] = pd.to_datetime(
    df["Date"],
    dayfirst=True,
    errors="coerce"
)
```

```python
#Check invalid dates:
invalid_dates = df["Date"].isna().sum()

print("Invalid Dates :", invalid_dates)
```

```python
#Clean the Amount Column
df["Amount"] = (
    df["Amount"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace("Rs.", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)
```

```python
df["Amount"] = pd.to_numeric(
    df["Amount"],
    errors="coerce"
)
```

```python
invalid_amounts = df["Amount"].isna().sum()

print("Invalid Amounts :", invalid_amounts)
```

```python
#Standardize Type Column
df["Type"] = df["Type"].astype(str).str.lower()

df["Type"] = df["Type"].replace({
    "dr": "debit",
    "debit": "debit",
    "cr": "credit",
    "credit": "credit"
})
```

```python
#Check Data Types
print(df.dtypes)
```

```python
#Final Parser Output
print("=" * 60)
print("TRANSACTION PARSER SUMMARY")
print("=" * 60)

print(f"Parsed Transactions      : {len(df)}")
print(f"Duplicates Removed       : {duplicate_rows}")
print(f"Invalid Dates            : {invalid_dates}")
print(f"Invalid Amounts          : {invalid_amounts}")

print("=" * 60)
```

```python
#TASK - 2
#View Unique Descriptions
print(df["Description"].unique())
```

```python
#Create Vendor Dictionary
def find_vendor(description):

    description = str(description).upper()

    if "SWIGGY" in description or "BUNDL" in description:
        return "Swiggy"

    elif "ZOMATO" in description:
        return "Zomato"

    elif "ZEPTO" in description:
        return "Zepto"

    elif "BLINKIT" in description:
        return "Blinkit"

    elif "AMAZON" in description or "AMZN" in description:
        return "Amazon"

    elif "FLIPKART" in description:
        return "Flipkart"

    elif "MYNTRA" in description:
        return "Myntra"

    elif "AJIO" in description:
        return "Ajio"

    elif "UBER" in description:
        return "Uber"

    elif "OLA" in description:
        return "Ola"

    elif "RAPIDO" in description:
        return "Rapido"

    elif "ZERODHA" in description:
        return "Zerodha"

    elif "GROWW" in description:
        return "Groww"

    elif "NETFLIX" in description:
        return "Netflix"

    elif "SPOTIFY" in description:
        return "Spotify"

    elif "BOOKMYSHOW" in description:
        return "BookMyShow"

    elif "ATM" in description:
        return "Cash Withdrawal"

    elif "UPI-" in description and "@" in description:
        return "P2P Transfer"

    else:
        return "Uncategorised"
```

```python
#Apply Function
df["vendor_clean"] = df["Description"].apply(find_vendor)
```

```python
#Check Output
print(df[["Description","vendor_clean"]].head(20))
```

```python
#Number of Vendors
print("Total Vendors")

print(df["vendor_clean"].nunique())
```

```python
#Uncategorised Vendors
uncat = df[df["vendor_clean"] == "Uncategorised"]

print(uncat["Description"])
```

```python
#TASK - 3
#Create Category Function
def find_category(vendor):

    if vendor == "Swiggy" or vendor == "Zomato":
        return "Food Delivery"

    elif vendor == "Zepto" or vendor == "Blinkit":
        return "Quick Commerce"

    elif vendor == "Amazon" or vendor == "Flipkart" or vendor == "Myntra" or vendor == "Ajio":
        return "E-commerce"

    elif vendor == "Uber" or vendor == "Ola" or vendor == "Rapido":
        return "Transport"

    elif vendor == "Zerodha" or vendor == "Groww":
        return "Investments"

    elif vendor == "Netflix" or vendor == "Spotify":
        return "Subscriptions"

    elif vendor == "BookMyShow":
        return "Entertainment"

    elif vendor == "Cash Withdrawal":
        return "Cash Withdrawal"

    elif vendor == "P2P Transfer":
        return "Personal Transfer"

    else:
        return "Others"
```

```python
#Create Category Column
df["category"] = df["vendor_clean"].apply(find_category)
```

```python
#Extract Hour from Time
df["Hour"] = df["Time"].str[:2].astype(int)

print(df[["Time","Hour"]].head())
```

```python
#Create Time Period
def find_time(hour):

    if hour >= 5 and hour < 12:
        return "Morning"

    elif hour >= 12 and hour < 17:
        return "Afternoon"

    elif hour >= 17 and hour < 21:
        return "Evening"

    else:
        return "Night"
```

```python
#Create New Column
df["Time_Period"] = df["Hour"].apply(find_time)
```

```python
#Display Output
print(df[["Hour","Time_Period"]].head())
```

```python
#Count Transactions in Each Time Period
print(df["Time_Period"].value_counts())
```

```python
#Spending by Time Period
time_spending = df.groupby("Time_Period")["Amount"].sum()

print()

print("Spending by Time Period")

print(time_spending)
```

```python
#Calculate Mean
category_mean = df.groupby("category")["Amount"].transform("mean")

print(category_mean.head())
```

```python
#Calculate Standard Deviation
category_std = df.groupby("category")["Amount"].transform("std")

print(category_std.head())
```

```python
#Calculate Z-Score
df["z_score"] = (df["Amount"] - category_mean) / category_std

print(df[["Amount","category","z_score"]].head())
```

```python
#Find Anomalies
anomaly = df[df["z_score"] > 2]
```

```python
#Display Top Anomalies
print(anomaly[["Date",
               "vendor_clean",
               "category",
               "Amount",
               "z_score"]])
```

```python
#Sort by Highest Z-Score
anomaly = anomaly.sort_values(
    by="z_score",
    ascending=False
)

print(anomaly.head(10))
```

```python
#Calculate Percentages
food_percent = ((food + restaurant + cafe) / total_debit) * 100

quick_percent = (quick / total_debit) * 100

ecommerce_percent = (ecommerce / total_debit) * 100

investment_percent = (investment / total_debit) * 100
```

```python
#Calculate Late Night Food Orders
food_order = df[df["category"] == "Food Delivery"]

late_night = food_order[
    (food_order["Hour"] >= 21) |
    (food_order["Hour"] <= 2)
]

late_percent = (len(late_night) / len(food_order)) * 100
```
